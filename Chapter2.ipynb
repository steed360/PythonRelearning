```python
import os

fileDir = 'C:\\Users\\John\\m1-100k_2\ml-100k'
os.listdir (fileDir)

###########################################################
# Read in user table
###########################################################

unames = ['user_id', 'age', 'gender' , 'occupation', 'zip']
df_users = pd.read_table ( os.path.join ( fileDir, 'u.user'), sep="|", header=None, names=unames ) 
#df_users


###########################################################
# Read in movie table
###########################################################
#movie id, 
m_names = "movie title | release date | video release date |" + \
              "IMDb URL | unknown | Action | Adventure | Animation |" + \
              "Children's | Comedy | Crime | Documentary | Drama | Fantasy |" + \
              "Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi |" + \
              "Thriller | War | Western "

i = m_names.split ('|')
movieColNames = [j.strip() for j in i ]


df_movies = pd.read_table ( os.path.join ( fileDir, 'u.item'), sep="|",encoding = "ISO-8859-1" , header=None, names =movieColNames)
#df_movies

df_movies['item_id'] = df_movies.index


###########################################################
# Read in recommendations table
###########################################################

r_names = 'user_id', 'item_id','rating', 'timestamp'

df_recs = pd.read_table  ( os.path.join ( fileDir, 'u.data'), 
                                sep="\t",encoding = "ISO-8859-1" , header=None, names =r_names  )
df_recs.head()





```

```python
# Merge all of the data into one table

df_data = pd.merge ( 
    pd.merge ( df_recs, df_users  , on = ['user_id', 'user_id'], how='left'), 
          df_movies, 
          on = ['item_id', 'item_id'], 
          how='left' )
#df_data.head()
```

```python
# Data manipulation gympastics
# a) mean_ratings : has rating for two columns M and F
#  - indexed by movie name
# b) df_titles_with_num_ratings is a subset of mean_ratings, only including movies with lots of ratings


# Aggregate the ratings: mean rating for each file grouped by gender...
mean_ratings= df_data.pivot_table ( 'rating',  columns = 'gender', index = 'movie title', aggfunc = 'mean')

# Find the number of ratings per film and then the names of films with many ratings.
# The resulting 
df_titles_with_num_ratings = df_data.groupby ('movie title').size()
df_titles_with_many_ratings [ df_titles_with_num_ratings >= 250]

idx_df_titles_with_many_ratings = df_titles_with_many_ratings.index


# Filter mean_ratings leaving those titles with many ratings.
mean_ratings [ mean_ratings.index.isin (idx_df_titles_with_many_ratings )]
```

```python

# Measuring Rating Disagreement
# Find movies that are most divisive between male and female viewers.

# First flag  NaN values rows as we want to ignore these.
# Pandas doesn't seem to do the 'right thing' like R does.


import numpy as np


# Add a column to mean_ratings : diff_in_means
 
mean_ratings['diff_in_means'] = mean_ratings.M - mean_ratings.F

# Large difference where the films were  preferred by Women
mean_ratings[ ~  mean_ratings.diff_in_means.isna()].sort_values (  by ='diff_in_means')[0:20]





```

```python
# Small differences where the files are preferred by Men
# To reverse using slicing with -1 step and then take the top 15 rows

mean_ratings[ ~  mean_ratings.diff_in_means.isna()].sort_values (  by ='diff_in_means') [:0:-1][0:16]
```

```python
# Films that elicit most disagrement between men and women, regardless of gender
# To calculate this go back to the original dataframe and find the standard deviation.

rating_std_by_title = df_data.groupby('movie title')['rating'].std()

# Order the series by value in descendig order.

filmsCausingDisagreement= rating_std_by_title[ ~rating_std_by_title.isna()  ] .sort_values (ascending=False) [ 0:20]
filmsCausingDisagreement


```
